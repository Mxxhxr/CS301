```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Load the dataset
data = pd.read_csv('PJ2.csv')
```

```python
#Explaining Dataset
'''
The data set is showcasing an amazon product and its reviews and their stats such as if it was a verified purchase or not,
what the star rating is for the product, the customer's id. etc. 
C = Categorical
N = Numerical
There are 16 columns
marketplace = tells you what marketplace the product was purchased in (C)
star_rating = tells us what rating the reviewing customer left (N)
helpful_votes = shows how many people upvoted the comment (N)
total_votes = shows how many people in general reacted with the review (N)
vine = this tells if it the review is a part of the amazon vine program (C but boolean)
verified_purchase = it tells you if the reviewer who purchased was verified by amazon (C but boolean)
customer_id = unique id of the customer who purchased the product (N)
review_id = the unique id of the review to distinguish every single review (C)
product_id = the unique id of the product (C)
product_parent = i think this is the unique id of the seller's profile (N)
product_title = the name of hte product (N)
product_category = tells what circle of products it falls under (C)
review_headline = tells you the first words of the review (C)
review_body = contains the contents of the review (C)
review_date = the date of the review (N)
sentiment = tells the connotation of the review (N)

The models we have used to train the model are deep neural network and decion tree regression. 
The highlighted categories we chose for decision tree regression are helpful_votes, total_votes, verified_purchase, and product category.
These will be trained to help find a pattern for star_rating.

The highlighted categories for the deep neural network are star_rating and total_votes with star_rating being the target variable. 

To clean up the data, we dropped all null values just in case and replaced them with the average of that column. Also we used one hot encoding and 
label encoding. We used one hot encoding to transform the product_category column and we used label encoding to transform the verified_purchase column.
'''
```

```python
# Count the occurrences of each star rating
star_rating_counts = data['star_rating'].value_counts().sort_index()

# Bar chart
plt.figure(figsize=(10, 6))
star_rating_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Create a joint plot
plt.figure(figsize=(10, 8))
sns.jointplot(x='total_votes', y='star_rating', data=data, alpha=0.5)
plt.xlabel('Total Votes')
plt.ylabel('Star Rating')
plt.grid(True)
plt.show()
```

```python
# Assuming 'star_rating' is a column in your DataFrame 'data'
plt.figure(figsize=(10, 6))
plt.scatter(data['helpful_votes'], data['total_votes'], c=data['star_rating'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Star Rating')
plt.xlabel('Helpful Votes')
plt.ylabel('Total Votes')
plt.title('Scatter Plot: Helpful Votes vs Total Votes')
plt.grid(True)
plt.show()
```
