## Task 3
**Q.3  Write python code using Matplotlib to create a line plot with multiple lines (for data use trigometric functions/ you can use any other dataset as well), labels, title, and a legend. (20 pts)**

```python
# Your code goes here
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.linspace(0, math.sqrt(50*3.14), 100)

xcos = np.cos(x)
xsin = np.sin(x)
xtan = np.tan(x)

fig, ax = plt.subplots()

ax.plot(x, xcos, label='cos(x)')
ax.plot(x, xsin, label='sin(x)')
ax.plot(x, xtan, label='tan(x)')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Some trig operations on x")

ax.legend()

plt.show()

```

## Task 4
**Q.4 Write python code using Matplotlib to create a scatter plot for visualizing a dataset with two numerical variables, labels, title, and a legend. (generate sample data using 'random'/you can use any other dataset as well)    ( 20 pts)**

```python
#  your code goes here 
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(40)
var1 = np.random.rand(100)
var2 = np.random.rand(100)
var3 = (np.random.rand(100) * 401.93) + 54


plt.scatter(var1, var2, s=var3, alpha=.8, label='Random Data')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with random data')


plt.legend()

plt.show()

```

## Task 5
**Q.5 Write a Python code using Matplotlib to generate a histogram (generate sample data using 'random'/you can use any other dataset as well)with labels, title, and a legend.  ( 20 pts)**

```python
#  your code goes here 
import matplotlib.pyplot as plt
import numpy as np

var1 = np.random.normal(100, 9, 200)


plt.hist(var1, label='Some Data')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Histogram w/ random data')


plt.legend()

plt.show()


```
