## What is Matplotlib?

Matplotlib is a Python library for creating **static, animated, and interactive visualizations**, commonly used for plotting 2D and 3D graphs and charts. It generates graphs by first creating a **figure**, representing the overall canvas, and **axes**, which are individual subplots within the figure. Users then use various plotting functions to input data and visualize it on the specified axes. The library offers extensive customization options, allowing users to control colors, styles, labels, and other plot features.

## Anatomy of a Figure

image.png

### Figure
In Matplotlib, a "figure" refers to the overall window or page that everything is drawn on. It can contain multiple axes (individual plots), titles, labels, and other elements. When you create a plot using Matplotlib, you typically start by creating a figure and then adding one or more axes to it.

### Axes
In Matplotlib, an "axes" is an individual plot within a figure. A figure can contain one or more axes, and each axis can have its own set of data, labels, and other graphical elements. The terms "axes" and "subplot" are often used interchangeably in Matplotlib. Following are some important components of an axes:

1. **Title:**
   - The title is a descriptive text placed at the top of the plot, providing an overall description of what the plot represents.

2. **X-axis:**
   - The x-axis is the horizontal line on the plot where data points are plotted. It represents the independent variable in the context of the plot.

3. **Y-axis:**
   - The y-axis is the vertical line on the plot where data points are plotted. It represents the dependent variable in the context of the plot.


4. **X-axis Label:**
   - The x-axis label is a description placed along the x-axis, providing information about the quantity or variable being represented on the horizontal axis.

5. **Y-axis Label:**
   - The y-axis label is a description placed along the y-axis, providing information about the quantity or variable being represented on the vertical axis.

6. **Ticks:**
   - Ticks are the marks on the x-axis and y-axis that represent specific data values. They help in understanding the scale of the plot and locating specific points.

7. **Tick Labels:**
   - Tick labels are the text labels associated with the ticks on the x-axis and y-axis. They provide the actual values corresponding to the tick marks and help in interpreting the data.

8. **Markers:**
   - Markers are symbols or points on the plot that indicate individual data points. They are useful in scatter plots and line plots to visually highlight the position of data along both the x-axis and y-axis.

9. **Legend:**
   - A legend is an area describing the elements of the plot. It provides an explanation for the meaning of the different colors, lines, or markers used in the plot. Legends are particularly useful when a plot contains multiple datasets or when different features of the data are represented with distinct visual styles.

### The explicit and the implicit Methods
There are two methods for generating plots in Matplotlib: the **implicit method** and the **explicit method**.

1. **Implicit Method:**
   - The implicit method involves letting Matplotlib handle the creation and management of figures and axes automatically.
     ```python
     import matplotlib.pyplot as plt
     plt.plot([1, 2, 3], [4, 5, 6])
     plt.show()
     ```
   - In this example, both the figure and axis are created implicitly by the `plt.plot()` function. Matplotlib automatically handles the creation of a figure and an axis if they don't exist.

2. **Explicit Method:**
  - The explicit method involves explicitly creating and managing the figure and axes objects.
    
    ```python
    import matplotlib.pyplot as plt
    fig, axes = plt.subplots(nrows=2, ncols=2)
    axes[0,0].plot([1, 2, 3], [4, 5, 6])
    plt.show()
     ```


````python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,1].plot([1, 2, 3], [4, 5, 6])
plt.show()
````

## Colors, Markers, and Line Styles

When creating plots with Matplotlib, you can customize the appearance of your plots using colors, markers, and line styles.

In the following example:
- `color` parameter sets the line color.
- `linestyle` parameter sets the line style ('-', '--', ':', etc.). Emtpy string will only show the markers.
- `marker` parameter sets the marker style ('o', 's', '^', etc.).
[marker api](https://matplotlib.org/stable/api/markers_api.html)

````python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5, 6]
y = [3, 5, 7, 9, 11, 14]

# Create a figure with 4 subplots
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(12, 2))

axes[0].plot(x, y, color='green', linestyle='--',marker="o") # Plot on the first subplot
axes[1].plot(x, y, 'rs-') # Plot on the second subplot

axes[2].plot(x, y, color='indigo', marker="o",linestyle="") # Plot on the third subplot
axes[3].plot(x, y,'bo:') # Plot on the fourth subplot
axes[4].plot(x,y, color='orange', linestyle="--", marker='x')

plt.show()
````

## Adjusting Spacing Around Subplots

You can use the `subplots_adjust or tight_layout` function to adjust spacing around subplots.

- The `subplots_adjust` function in Matplotlib is used to adjust the spacing between subplots. It allows you to control the spacing (e.g., left, right, top, bottom) between subplots and the overall figure.

- The **`tight_layout`** function in Matplotlib is used to automatically adjust subplot parameters to fit the figure area.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure with 2x2 subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))

axes[0, 0].plot(x, y,color='blue',marker="+",linestyle="--")
axes[0, 1].plot(x, y,'go', color='orange', marker='o', linestyle='dashdot')
axes[1, 0].plot(x, y,'ko')
axes[1, 1].plot(x, y,'ro')
axes[1, 1].plot(x, y,'ro')

axes[0,0].set_title('Top Left')
axes[0,1].set_title('Top Right')
axes[1,0].set_title('Bottom Left')
axes[1,1].set_title('Bottom Right')

plt.subplots_adjust(left=.1, right=0.9, top=50, bottom=0.0, wspace=0.3, hspace=0.5) #keep all values between 0 and 1
# left, right, top, and bottom parameters control the spacing of the subplot grid with respect to the figure edges. wspace and hspace parameters control the width and height reserved for space between subplots, respectively.

plt.tight_layout()

plt.show()
````

## Sharing Axis

When creating subplots using Matplotlib, the `sharex` and `sharey` parameters can be used to share the x-axis and y-axis, respectively, among multiple subplots. This is useful when you want to synchronize the axes of different subplots.

In the following example, `sharex=True` is used when creating the subplots, which means that all subplots in the same column share the same x-axis. This ensures that the x-axis scales are the same across all subplots, making it easier to compare different datasets.

You can similarly use `sharey=True` to share the y-axis among subplots in the same row. Adjust these parameters based on your specific requirements for sharing axes among subplots.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure with 2x2 subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6),sharex=True, sharey=True)

# Plot on the first subplot (upper-left)
axes[0, 0].plot(x, y,'bo')
axes[0, 0].set_title('Plot 1')

# Plot on the second subplot (upper-right)
axes[0, 1].plot(x, y,'go')
axes[0, 1].set_title('Plot 2')

# Plot on the third subplot (lower-left)
axes[1, 0].plot(x, y,'ko')
axes[1, 0].set_title('Plot 3')

# Plot on the fourth subplot (lower-right)
axes[1, 1].plot(x, y,'ro')
axes[1, 1].set_title('Plot 4')

# Use tight_layout to automatically adjust subplot parameters
plt.tight_layout()

# Show the plot
plt.show()
````

## Title, Axis Labels, Ticks, and Ticklabels

````python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]

# Create a Figure and Axes object
fig, ax = plt.subplots(figsize=(5, 4))


ax.plot(x, y, 'go')

# Set title
ax.set_title('My Plot Title', fontsize=14)

# Set x and y axis labels
ax.set_xlabel('X-axis Label', fontsize=12)
ax.set_ylabel('Y-axis Label', fontsize=12)

# Set x and y axis ticks
ax.set_xticks([1,3,4])
ax.set_yticks(y)

# Set x and y axis tick labels
ax.set_xticklabels(['One', 'Three', 'four'], fontsize=10)

ax.set_yticklabels(y, fontsize=10)

ax.legend()
# Display the plot
plt.show()
````

## Adding Legends

Legends in Matplotlib can be used to label different elements of a plot.

In the following example:
- We create a single set of x and y axes using `plt.subplots()` and store the Figure and Axes objects in `fig` and `ax` variables.
- Different plots are added to the same set of axes using `ax.plot()`.
- Each `plot` call has a `label` parameter, which is used to identify the plots in the legend.
- The `ax.legend()` call adds a legend to the plot using the labels specified in the `plot` calls.

````python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y1 = [3, 5, 7, 9, 11]
y2 = [2, 4, 6, 8, 10]
y3 = [1, 3, 5, 7, 9]

fig, ax = plt.subplots(figsize=(6, 3))

ax.plot(x, y1, color='blue', linestyle='-', label='exp1')
ax.plot(x, y2, color='red', linestyle='--', marker='s', label='exp2')
ax.plot(x, y3, "go-", label='exp3')

ax.legend(loc=0)

plt.show()
````

## Annotation in Matplotlib

Matplotlib provides the `annotate` function to add annotations to your plots. Here's a simple example:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create subplots
fig, ax = plt.subplots()

ax.scatter(x, y, color='red', label='Scatter Data')
ax.set_title('Subplot 2')

ax.annotate('This point is interesting!',
                xy=(2, 4),  # Point to annotate (x, y)
                xytext=(2.5, 5),  # Position of the text label
                arrowprops=dict(edgecolor='blue', arrowstyle='->'))

# Show the plot
plt.show()

```

In this example, we have a simple plot of data points, and we use `annotate` to add a text annotation to a specific point on the plot. Adjust the `xy` and `xytext` parameters as needed for your specific use case. The `arrowprops` argument is optional and is used to add an arrow pointing to the annotated point.

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create subplots
fig, ax = plt.subplots()

ax.plot(x, y, color='red', label='Scatter Data', linestyle='', marker='o')
ax.set_title('Subplot 2')

ax.annotate('This point is interesting!',
                xy=(2, 4),  # Point to annotate (x, y)
                xytext=(2, 8),  # Position of the text label
                arrowprops=dict(edgecolor='blue', arrowstyle='->'))

ax.annotate('This point is (3,6)!',
                xy=(3, 6),  # Point to annotate (x, y)
                xytext=(3, 5),  # Position of the text label
                arrowprops=dict(edgecolor='blue', arrowstyle='->'))
# Show the plot
plt.show()
````

## Customizing Colors

In Matplotlib, you can customize the colors of various components of a plot, including lines, spines (the borders of the plot), markers, backgrounds, and text. Here's an example demonstrating how to change the colors of different plot components:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))

# Line plot with a blue line
ax.plot(x, y, label='Sin(x)', color='blue')

# Scatter plot with green circles
ax.scatter(x, y, label='Data Points', color='green', marker='o')

# Background color of the plot area
ax.set_facecolor('lightgray')

# Title color
ax.title.set_color('red')
ax.set_title("Customizing Colors")

# Customize tick colors
ax.tick_params(axis='both', colors='darkblue')

# Legend text color
legend = ax.legend()
for text in legend.get_texts():
    text.set_color('orange')

# Grid color
ax.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Add a text annotation with a different color
ax.annotate('Max Value', xy=(np.pi / 2, 1), xytext=(np.pi / 2, 1.5),
            arrowprops=dict(edgecolor='red', arrowstyle='->'), fontsize=10, color='red')

# Customize spine color
ax.spines['bottom'].set_color('darkblue')
ax.spines['top'].set_color('darkblue')
ax.spines['right'].set_color('darkblue')
ax.spines['left'].set_color('darkblue')

# Customize axis label color
ax.set_xlabel('X-axis', color='darkblue')
ax.set_ylabel('Y-axis', color='darkblue')

# Show the plot
plt.show()

```

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))

# Line plot with a blue line
ax.plot(x, y, label='Sin(x)', color='blue')

# Scatter plot with green circles
ax.scatter(x, y, label='Data Points', color='green', marker='o')

# Background color of the plot area
ax.set_facecolor('lightgray')

# Title color
ax.title.set_color('red')
ax.set_title("Customizing Colors")

# Customize tick colors
ax.tick_params(axis='both', colors='darkblue')

# Legend text color
legend = ax.legend()
for text in legend.get_texts():
    text.set_color('orange')

# Grid color
ax.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Add a text annotation with a different color
ax.annotate('Max Value', xy=(np.pi / 2, 1), xytext=(np.pi / 2, 1.5),
            arrowprops=dict(edgecolor='red', arrowstyle='->'), fontsize=10, color='red')

# Customize spine color
ax.spines['bottom'].set_color('darkblue')
ax.spines['top'].set_color('darkblue')
ax.spines['right'].set_color('darkblue')
ax.spines['left'].set_color('darkblue')

ax.spines['bottom'].set_linestyle('-')
ax.spines['top'].set_linestyle('-')
ax.spines['right'].set_linestyle('-')
ax.spines['left'].set_linestyle('-')

# Customize axis label color
ax.set_xlabel('X-axis', color='darkblue')
ax.set_ylabel('Y-axis', color='darkblue')

# Show the plot
plt.show()
````

## Saving Plots to File

In Matplotlib, you can save a figure to a file using the `savefig` function. This function allows you to save the current figure to a file in various formats, such as PNG, PDF, SVG, and more.

In the following example, after creating the figure and plotting the data, the `savefig` method is used to save the figure. The argument passed to `savefig` is the filename, and the file format is determined by the file extension.

Remember to call `plt.show()` after saving the figure if you want to display it in the interactive mode.

Adjust the filename and format according to your needs.

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create subplots
fig, ax = plt.subplots()

ax.scatter(x, y, color='red', label='Scatter Data')
ax.set_title('Subplot 2')

ax.annotate('This point is interesting!',
                xy=(2, 4),  # Point to annotate (x, y)
                xytext=(2.5, 5),  # Position of the text label
                arrowprops=dict(edgecolor='blue', arrowstyle='->'))

# Save the plot to a file (e.g., PNG)
plt.savefig('figure.png')

# Show the plot
plt.show()
````

## Gridspec

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Create a 2x2 grid of subplots with different sizes
fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2, width_ratios=[2, 1], height_ratios=[1, 2])

# Create subplots using gridspec
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, :])

# Plot some data in each subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [1, 2, 3])
ax3.plot([1, 2, 3], [7, 8, 9])

# Set titles for each subplot
ax1.set_title('Subplot 1')
ax2.set_title('Subplot 2')
ax3.set_title('Subplot 3')

# Set title for the entire figure
plt.suptitle('Gridspec')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

```
Here, ax3 spans the entire second row (gs[1, :]).
Also, the width_ratios parameter [2, 1] indicates that the first column will be twice as wide as the second column. Similarly, the height_ratios parameter [1, 2] indicates that the first row will be half as tall as the second row.

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Create a 2x2 grid of subplots with different sizes
fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2, width_ratios=[2, 1], height_ratios=[1, 2]) 

# Create subplots using gridspec
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, :])

# Plot some data in each subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([1, 2, 3], [1, 2, 3])
ax3.plot([1, 2, 3], [7, 8, 9])

# Set titles for each subplot
ax1.set_title('Subplot 1')
ax2.set_title('Subplot 2')
ax3.set_title('Subplot 3')

# Set title for the entire figure
plt.suptitle('Gridspec')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````
